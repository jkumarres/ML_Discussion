# Finding 1D minima using gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

## The function and its gradient

```python
x, y= symbols("x y")

z = -100 * x**2 + x*y - 1 * x**3  + x**4
dydx = diff( y )

print( dydx )
```

```python
NumberGrid = 100

xValue = np.linspace(-10,10,NumberGrid)
yValue = np.zeros( NumberGrid )

for i in range(NumberGrid) :
    yValue[i] = y.subs( x , xValue[i] )

plt.plot( xValue , yValue )
plt.show()
```

## Performing gradient descent

Parameters

```python
# Initial value
x0, y0 = -10, 1

# Hyperparameters
MaxIterations = 200
LearningRate = 0.001
```

```python
xi = x0
xHistory = np.zeros( MaxIterations )

for i in range( MaxIterations ):
    dydxValue = dydx.subs( x , xi )

    xi = xi - LearningRate * (dydxValue + np.random.rand(1) )
    xHistory[i] = xi

plt.plot( xHistory )
plt.title( "Evolution of the initial value" )
plt.show()

plt.plot( xValue , yValue )
plt.plot( x0 , y.subs(x,x0) , '*b' )
plt.plot( xi , y.subs(x,xi) , 'or' )
plt.title( "Initial and final vaule" )
plt.show()
```
